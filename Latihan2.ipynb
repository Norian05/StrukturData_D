Definisi List

```python
list = []
```

Input 10 data menggunakan perulangan 'For'

```python
for i in range(10):
    numer = int(input("Input angka : "))
    list.append(numer)
```

Menampilkan data list, nilai tertinggi, terendah dan rata-rata

```python
print("List Angka:", list)
print("Nilai Tertinggi: ", max(list))
print("Nilai Terendah : ", min(list))
print("Nilai Rata-rata:", sum(list)/len(list))
```

Tampilan Grafik Bar

```python
import matplotlib.pyplot as plt
plt.bar(range(len(list)), list)
plt.title("Grafik")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.show()
```
